```python
from datetime import datetime as dt
```

```python
netflix["computer_date"].fillna(dt(2020, 1,1), inplace=True)  # na or Na or NaN = Not a Number or Null Value
```

Spliting Columns

```python
netflix["date_added"].str.split(", ", expand=False) # we don't want this. NOTE - False
```

```python
netflix["date_added"].str.split(", ", expand=True) # we want this. NOTE - True
```

```python
netflix[["Date Part 1", "Date Part 2"]] = netflix["date_added"].str.split(", ", expand=True)
netflix
```

Applying a Function to Every Row

```python
the_adder = lambda x: x + 210101    # just performing addition NOTED
the_adder(3535)
```

```python
netflix["genre_count"] = netflix["listed_in"].map(lambda x: len(x.split(", ")))
netflix
```

```python
netflix["nonsensical_columns"] = netflix.apply(lambda x: len(x["cast"].split(",")) + len(x["listed_in"].split(",")), axis = 1)
netflix
```

GroupBy, Aggregation, and Sorting Values

```python
netflix.groupby("country")["show_id"].count()
```

```python
netflix.groupby("country")["show_id"].count().reset_index()  # .rest_index(), turns it to a nice clear data frame
```

```python
netflix_movies_by_country = netflix.groupby("country")["show_id"].count().reset_index().sort_values(by="show_id", ascending=False)   # False and True changes the order in Show_id
netflix_movies_by_country
```

Joins & Unions

```python
second_dataset = pd.read_csv(pwd + "/netflix_titles_second.csv")
second_dataset
```

```python
new_dataset = pd.concat([netflix, second_dataset])  # pd.concat means concatdinate = Joining two things or staking 2 data frames together
new_dataset
```

```python
netflix_merged = pd.merge(left = new_dataset, right = netflix_movies_by_country, how ="inner", left_on = ["country"], right_on = ["country"]) # merge means join
netflix_merged
```

Pivote and Melt

```python
pivot_table = netflix.pivot_table(index="country", columns="type", values= "title", aggfunc='count', fill_value=0).reset_index()
pivot_table
```
